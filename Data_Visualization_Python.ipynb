```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--")

plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.legend(("sin", "cos"))
```

```python
fig = plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), "-", linewidth=4, color="red")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.legend(("sin"))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), "--")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.legend(("cos"))

```

### LINE PLOT

```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris["petal_length"].plot()  #pandas library provides to create plot so easy.
```

```python
sns.lineplot(data=iris, y = "petal_length", x = iris.index, hue="species")
```

### SCATTER PLOT
* İki sayısal grup arasındaki ilişkiyi dağınık noktalar şeklinde çizmek için kullanılır. Grafikte ki her nokta bir gözlemi temsil eder. X ekseni gözlemin bir özelliğini, Y ekseni diğer bir gözlemi ifade eder.
* Korelasyon hakkında bilgi verir. 
* Çok fazla veri olduğu durumlarda **overplotting** hatasıyla karşılaşılabilir
 1. Bu hatayı çözmek için nokta alanları küçültülebilir.
 2. Renk yoğunluklarıyla yapılabilir.

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.scatter(iris["sepal_length"], iris["sepal_width"]);
ax.set_title("Iris Dataset")
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
```

### BAR PLOT
* Toplam, ortalama, medyan gibi Aggregated Func.'ları kullanarak bir kategorik verinin sayısal olarak gruplandırmak.
* x ekseninde kategoriler, y ekseninde sayısal veriler olur.
* Bar plot(çubuk grafiğinde) ayrık verilerle çalışılırken, Histogramlarda sürekli veriler gösterilir.


```python
iris.groupby(by="species").mean()
```

```python
iris.groupby(by="species").max()
```

```python
iris.groupby(by="species").sum()["sepal_length"]
```

```python
data = iris.groupby(by="species").sum()["sepal_length"]
ax = sns.barplot(x=data.index, y = data.values)
ax.set_title("Sepal Length Means")
ax.set_ylabel("Mean")
```

### HISTOGRAM

* Sürekli verilerin belirli aralıklara bölünerek gösterilmesidir.

```python
sns.histplot(data = iris, x = "petal_length")
```

```python
sns.histplot(data = iris, x = "sepal_width", bins = 20)
```
