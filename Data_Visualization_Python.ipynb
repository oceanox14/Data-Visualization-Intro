```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--")

plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.legend(("sin", "cos"))
```

```python
fig = plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), "-", linewidth=4, color="red")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.legend(("sin"))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), "--")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.legend(("cos"))

```

### LINE PLOT

```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris["petal_length"].plot()  #pandas library provides to create plot so easy.
```

```python
sns.lineplot(data=iris, y = "petal_length", x = iris.index, hue="species")
```

### SCATTER PLOT
* ki saysal grup arasndaki ilikiyi dank noktalar eklinde izmek iin kullanlr. Grafikte ki her nokta bir gzlemi temsil eder. X ekseni gzlemin bir zelliini, Y ekseni dier bir gzlemi ifade eder.
* Korelasyon hakknda bilgi verir. 
* ok fazla veri olduu durumlarda **overplotting** hatasyla karlalabilir
 1. Bu hatay zmek iin nokta alanlar kltlebilir.
 2. Renk younluklaryla yaplabilir.

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.scatter(iris["sepal_length"], iris["sepal_width"]);
ax.set_title("Iris Dataset")
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
```

### BAR PLOT
* Toplam, ortalama, medyan gibi Aggregated Func.'lar kullanarak bir kategorik verinin saysal olarak gruplandrmak.
* x ekseninde kategoriler, y ekseninde saysal veriler olur.
* Bar plot(ubuk grafiinde) ayrk verilerle allrken, Histogramlarda srekli veriler gsterilir.


```python
iris.groupby(by="species").mean()
```

```python
iris.groupby(by="species").max()
```

```python
iris.groupby(by="species").sum()["sepal_length"]
```

```python
data = iris.groupby(by="species").sum()["sepal_length"]
ax = sns.barplot(x=data.index, y = data.values)
ax.set_title("Sepal Length Means")
ax.set_ylabel("Mean")
```

### HISTOGRAM

* Srekli verilerin belirli aralklara blnerek gsterilmesidir.

```python
sns.histplot(data = iris, x = "petal_length")
```

```python
sns.histplot(data = iris, x = "sepal_width", bins = 20)
```
